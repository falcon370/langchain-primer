```python
from langchain_openai import ChatOpenAI
from langchain.schema import HumanMessage

# Initialize the LLM
llm = ChatOpenAI(
    model='gpt-3.5-turbo',
    temperature=0.7
)

# Make your first call
response = llm.invoke([
    HumanMessage(content='Hello!')
])
print(response.content)
```
